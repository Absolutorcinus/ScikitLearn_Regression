```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([ 1.5,3.45,5.46,7.2,8.45,14,10.05,7.34,2.68,9.49,]).reshape(10,1)
y = np.array([8.034,18.099,30.378,39.737,42.57,73.951,52.818,38.973,17.373,50.426])

print(x.shape , y.shape)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x,y)
```

```python
Yhat=lm.predict(x)
Yhat[:]   
```

```python
(lm.intercept_)
```

```python
lm.coef_
```

```python
yhat =  lm.intercept_ + lm.coef_ * x
yhat
```
